````sas
/*Problem 7*/
data sba_subset;
    set sba;
    keep Name City State ApprovalFY NoEmp MIS_Status GrAppv SBA_Appv;
run;
````

````sas
proc sort data=sba_subset out=sba_subset_sorted;
    by ApprovalFY;
run;

proc print data=sba_subset_sorted(obs=10);
run;
````

````sas
/*Problem 8*/
data sba_subset_renamed;
    set sba_subset;
    length Business_Size $10;
    rename NoEmp=Total_Employees
           ApprovalFY=Approval_Year
           MIS_Status=Loan_Status
           GrAppv=Gross_Approved_Amount
           SBA_Appv=SBA_Approved_Amount;
    SBA_Guaranteed_Proportion=SBA_Appv / GrAppv;
    if NoEmp <= 1 then
        Business_Size = "Small";
    if NoEmp >= 2 and Total_Employees <=20 then
        Business_Size = "Medium";
    if NoEmp > 20 then
        Business_Size = "Large";
    if State = "DE" and City = "DOVER" or City = "Dover";
run;
````

````sas
proc print data=sba_subset_renamed(obs=10);
run;
````

````sas
/*Problem 9*/
proc means data=sba_subset_renamed mean median mode var std noprint;
    class Approval_Year Business_Size;
    var Gross_Approved_Amount;
    output out=basicstats
        mean = mean_GAP median = median_GAP
        mode = mode_GAP var = var_GAP std = std_GAP;
run;
proc print data=basicstats(obs=15);
run;
````

````sas
/*Problem 10*/
proc sort data=sba_subset_renamed out=sba_subset_grouped2;
    by State Loan_Status;
run;

data props;
    set sba_subset_grouped2;
    by State Loan_status;
    if nmiss(of _numeric_) + cmiss(of _character_) > 0 then delete;
    mean_SBA_GP = mean(SBA_Guaranteed_Proportion);
run;

proc sort data=props;
    by mean_SBA_GP;
run;

proc print data=props(obs = 10);
run;
    
````

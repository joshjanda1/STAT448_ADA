**#18** Using SAS, use logistic regression to predict whether a baby is born prematurely or not (OEGest_R3). You must compare 2 sets of predictors to seek a better prediction. This problem requires the use of both **birthtrn** (building the model) and **birthtst** (testing how well it performs). 

Describe the results of the better model and how good it is at predicting if baby is born prematurely. (SAS, GLM - Logistic Regression, Data Visualization, Markdown)

My prior belief for this model is that anything related to the father should only have a very small effect, if any, on whether or not the baby is born prematurely or not. So, I will create two models. The first model will contain father related variables, and the second will not. These models will be scored on the test data.

```sas
proc logistic data=birthtrn outmodel=logistic_model1;
 class MRACE6 MEDUC FRACE6 FEDUC PRIORLIVE PRIORDEAD ME_PRES DMETH_REC ATTEND PAY_REC DPLURAL riskfactorbin / param=ref ref=first;
 model OEGEST_R3(event="1") = pwgt_r dwgt_r wtgain apgar10 previs m_ht_in fagecomb mager
                    mrace6 meduc frace6 feduc priorlive priordead me_pres
                    dmeth_rec attend pay_rec apgar5 dplural riskfactorbin;
 score data=birthtst out=preds1;
 ods select rsquare;
run;
proc logistic data=birthtrn outmodel=logistic_model2;
 class MRACE6 MEDUC PRIORLIVE PRIORDEAD ME_PRES DMETH_REC ATTEND PAY_REC DPLURAL riskfactorbin / param=ref ref=first;
 model OEGEST_R3(event="1") = pwgt_r dwgt_r wtgain apgar10 previs m_ht_in mager
                    mrace6 meduc  priorlive priordead me_pres
                    dmeth_rec attend pay_rec apgar5 dplural riskfactorbin;
 score data=birthtst out=preds2;
 ods select rsquare;
run;
```

```sas
data preds1_class;
    set preds1;
    if p_1 >= 0.5 then pred = 1;
    else pred = 2;
    keep pred;
run;
data preds2_class;
    set preds2;
    if p_1 >= 0.5 then pred = 1;
    else pred = 2;
    keep pred;
run;

proc sort data=birthtst out=birthtst_sorted(keep=OEGest_R3);
    by OEGest_R3;
run;
proc sort data=preds1_class;
    by pred;
run;
proc sort data=preds2_class;
    by pred;
run;

data tst_preds1;
    merge birthtst_sorted preds1_class;
    match = 0;
    if OEGest_R3 = pred then match = 1;
run;
data tst_preds2;
    merge birthtst_sorted preds2_class;
    match = 0;
    if OEGest_R3 = pred then match = 1;
run;
```

```sas
title 'Father Variables Included';
proc means data=tst_preds1 mean;
    var match;
run;
title 'Father Variables not Included';
proc means data=tst_preds2 mean;
    var match;
run;
title;
```

Looking at the output above of the mean accuracy rate on the test dataset, we get a better accuracy score from the model where father variables are not included. For included father variables, our accuracy rate is ~.9589 whereas when not including father variables our accuracy rate is ~.9592. This supports my hypothesis that father related information does not play a large role on whether or not a baby is born prematurely.
